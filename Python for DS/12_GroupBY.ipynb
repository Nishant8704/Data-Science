# GroupBy
It is used for grouping the data same in DataBase

It apply on categorical Data

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('DataSets/imdb-top-1000.csv')
movies.head(1)
```

```python
movies.shape
```

```python
genres = movies.groupby('Genre')
genres
```

```python
# applying builtin aggregation functions on groupby object
genres.mean(" ")
genres.sum()
genres.min()
```

```python
# find the top three genres by total earning
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# Find the genre with highest avg IMBD rating
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# find director with most poplatity
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# find the highest rated movie of each genre
movies.groupby('Genre')['IMDB_Rating'].max()
```

```python
# find number of movies done by each actor
movies['Star1'].value_counts()
# Another way
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
# no. of unique value in data
genres.nunique()
```

```python
# agg method
genres.sum(" ")
```

```python
# Different column differnet aggregate function
genres.agg({
    'Runtime':'mean',
    'IMDB_Rating':'mean',
    'No_of_Votes':'sum',
    'Gross':'sum',
    'Metascore':'mean'
})
```

```python
# PAssing List
genres.agg(['min','max','sum'])
```

```python
#lopping in a group
# find the highest rated movie of each genres
df = pd.DataFrame(columns = movies.columns)
for group,data in genres:
    df = df.append(data[data['IMDB_Rating'] == data['IMDB_Rating']])

df
```

```python
#split (Apply) combine
#apply - > builtin function
genres.apply(min)
```

```python
# find number of movies stating with A for each group
def foo(group):
    num = group['Series_Title'].str.startswith('A').sum()
    return num
```

```python
genres.apply(foo)
```

```python
# find ranking of each movies in the group according to IMDB_Rating
def rank_movie(group):
    group['Genre_Rank']=group['IMDB_Rating'].rank(ascending =False)
    return group
```

```python
genres.apply(rank_movie)
```

```python
# find the normalize rating group wise
def normalized_Rank(group):
    group['Normal_Rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group
```

```python
genres.apply(normalized_Rank)
```

Group OF MUltiple Column

```python
# groupby on multiple columns
duo  = movies.groupby(['Director','Star1'])
duo
# size
duo.size()
# get group
duo.get_group(('Abhishek Kapoor','Amit Sadh')) 
```

```python
# find the most earning actor -> director combo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# find the best(in term of metascore(avg)) actor->genre combo
actgen = movies.groupby(['Star1','Genre'])
actgen.size()
actgen['Metascore'].sum().sort_values(ascending=False).head()
```

```python
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False).head(1)
```

```python
# agg on multiple columns
duo.agg(['max','min'])
```

Ball by ball data of IPL

```python
ipl = pd.read_csv("E:\Data Science\Python for DS\DataSets\deliveries.csv")
ipl.shape
```

```python
ipl.head()
```

```python
# find the top 10 batman in term of run score
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with most number of sixes
six = ipl[ipl['batsman_runs']==6]
```

```python
six.groupby('batsman')['batsman'].count().sort_values(ascending=False)
```

```python
# find batsman with most number of 4's and 6's in last 5 over
temp_df = ipl[ipl['over']>15]
temp_df = temp_df[(temp_df['batsman_runs']==4) | (temp_df['batsman_runs']==6)]
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False)
```

```python
# find virat Kohli record against all teams
ipl.head()
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index().sort_values('batsman_runs',ascending=False)
```

```python
# create a function that can return the score of any columns
temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest_score(batsman):
    temp_df = ipl[ipl['batsman'] == batsman]
    return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
batsman = input("Enter the batsman name")
highest_score(batsman)
```
