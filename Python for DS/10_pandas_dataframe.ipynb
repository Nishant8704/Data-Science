## Filtering A DataFrame

```python
ipl.head(2)
```

```python
# find all the final winners
mask = ipl['MatchNumber'] == 'Final'
new_df = ipl[mask]
final_winner  =new_df[['Season','WinningTeam']]
# print(final_winner)

# Alternative
ipl[ipl['MatchNumber'] == 'Final'][['Season','WinningTeam']]
```

```python
# How many super over finishes have occures
ipl.head(2)
mask = ipl['SuperOver']== 'Y'
superOver = ipl[mask].shape[0]
superOver
```

```python
# how many matches has csk won in kolkata
ipl.head()
mask = ipl['City'] == 'Kolkata' 
city = ipl[mask]
city[city['WinningTeam']=='Chennai Super Kings']

# Alternative
ipl[(ipl['City']=='Kolkata') & (ipl['WinningTeam']=='Chennai Super Kings')]
```

```python
# toss winner is match winner in percentage
ipl.head(2)
ans  = (ipl[ipl['TossWinner']==ipl['WinningTeam']].shape[0]/ipl['TossWinner'].shape[0])*100

print("total toss winner is match winner ", ans)
```

```python
# movies with rating higher than 8 and votes>10000
movies.head(2)
# shape[0]   is used for print number of rows
movies[(movies['imdb_rating']>=8) & (movies['imdb_votes']>10000)].shape[0]
```

## Important DataFrame Functions

### optimization of DataType so the memory reduce   ---> using astype

```python
# astype    ---> Change the data type of a given columns
ipl.info()
```

```python
# Permanent changing of a dataType
ipl['ID'] =  ipl['ID'].astype('int32')
ipl.info()
```

```python
# changing the repeatable value into category
ipl['Season'] = ipl['Season'].astype('category')
ipl['Team1'] = ipl['Team1'].astype('category')
ipl['Team2'] = ipl['Team2'].astype('category')
ipl.info()
```
