```python
import nlpcloud
class NLPApp:

  def __init__(self):
    self.__database = {}
    self.__first_menu()
    self.__second_menu()

  def __first_menu(self):
    first_input= input("""
    Hi! How you would like to proceed?
    1. Not a Member? Register
    2. Already a Member? Login
    3. Galti se aagya kya? Exit
    """)

    if first_input =='1':
      self.__register()
    elif first_input =='2':
      self.__login()
    else:
      exit()


  def __second_menu(self):
    second_input  = input("""
    What you want to do here: 
    1. NER
    2. Language Detection
    3. Sentiment Analysis
    4. Logout
    """)

    if second_input =='1':
      self.__ner()
    elif second_input =='2':
      self.__language_detection()
    elif second_input =='3':
      self.__sentimental()
    else:
      exit()


  def __register(self):
    name = input("Enter your Name: ")
    email = input("Enter Email Address: ")
    password = input("Your Password: ")
    
    if email in self.__database:
      print("Email Already Exist")
    else:
      self.__database[email]= [name,password]
      print("Registeration Successful")
      print(self.__database)
      self.__first_menu()


  def __login(self):
    email  = input("Enter your Email: ")
    password = input("Enter Password: ")

    if email in self.__database:
      if self.__database[email][1]==password:
        print("You have login successfully")
      else:
        print("Wrong Password")
        self.__login()
        self.__second_menu()
    else:
      print("You are not registered:")
      self.__register()

  def __ner(self):
    para= input("Enter the paragraph: ")
    search_term= input("What would you like to searh:")

    client = nlpcloud.Client("finetuned gpt-neox-20b","2b58",gpu = True,lang ='en')
    response = client.entities(para,searched_entity=search_term)
    print(response)

obj  = NLPApp()
```
