```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

# 2-D Plot

```python
price = [40000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('Datasets\sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# label Title and axis
plt.plot(batsman['index'],batsman['V Kohli'],color ='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue')

plt.title('Rohit Sharma v/s Virat Kohli')
plt.xlabel('Year')
plt.ylabel('Run Scored')
```

```python
# Line Style
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle = 'dashed',linewidth = 3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=4)

plt.title('Rohit Sharma v/s Virat Kohli')
plt.xlabel('Year')
plt.ylabel('Run Scored')
```

```python
# pointing the important point
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle = 'dashed',linewidth = 3,marker='D',markersize =6)
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=2,marker='o',markersize=8)

plt.title('Rohit Sharma v/s Virat Kohli')
plt.xlabel('Year')
plt.ylabel('Run Scored')
```

```python
# legend --> Location
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle = 'dashed',linewidth = 3,marker='D',markersize =6,label= 'virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=2,marker='o',markersize=8,label='rohit')

plt.title('Rohit Sharma v/s Virat Kohli')
plt.xlabel('Year')
plt.ylabel('Run Scored')

# calling label using legends
plt.legend()
```

```python
# limmiting axes
price = [40000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)

plt.ylim(0,75000)
plt.xlim(2015,2018)
```

```python
# Grid
plt.plot(batsman['index'],batsman['V Kohli'],color ='green',linestyle = 'dashed',linewidth = 3,marker='D',markersize =6,label= 'virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='blue',linestyle='dotted',linewidth=2,marker='o',markersize=8,label='rohit')

plt.title('Rohit Sharma v/s Virat Kohli')
plt.xlabel('Year')
plt.ylabel('Run Scored')

# calling label using legends
plt.legend()
plt.grid()
plt.show()
```

# Scatter Plot
-> Special Case of Line Plot

* Bivariate Analysis
* Numerical V/S Numerical
* Use cases - Finding Correlation

```python
# simple example
x= np.linspace(-10,10,50)
x

y  = 10*x +3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(x,y)
```

```python
# scatter on dataset
df = pd.read_csv('Datasets/batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='red',marker='D')
plt.title('Average V/S Strike_Rate of Batsman')
plt.xlabel('Average')
plt.ylabel('Strike Rate')

plt.grid()
plt.show()
```
